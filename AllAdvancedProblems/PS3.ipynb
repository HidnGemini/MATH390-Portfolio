```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# part a
ks = [10,100,1000]
for k in ks:
    matrix = np.random.randn(k,k)
    matrix = matrix/math.sqrt(k)
    eigens = np.linalg.eigvals(matrix)
    for eigen in eigens:
        x = eigen.real
        y = eigen.imag
        plt.scatter(x,y)
    plt.show()
```

```python
# part b
for k in ks:
    matrix = np.random.randn(k,k)
    matrix = (matrix + np.transpose(matrix))
    matrix = matrix/math.sqrt(2*k)
    eigens = np.linalg.eigvals(matrix)
    plt.hist(eigens)
    plt.show()
```

```python
# part c
"""
The eigenvalues of a random k*k matrix tend towards a circle of radius root(k) on the real, complex plane as k increases.
The eigenvalues are much more likely to land closer to 0, causing a semicircle when plotted as a histogram.
"""
```
